```python
mask = df['Room number'].isna()
df['text value'] = np.where (mask, df['Date'], "")
df
```

```python
df['text value'] = np.where(mask, df['Date'], np.nan)
df
```

```python
df['text value'].fillna(method= 'bfill',inplace= True)
df
```
